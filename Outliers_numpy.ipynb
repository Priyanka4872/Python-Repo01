```python
import numpy as np
sales=[15,18,20,22,25,28,30,32,35,38,40,42,45,48,50,52]
Q1=np.percentile(sales,25)
Q2=np.percentile(sales,50)
Q3=np.percentile(sales,75)
IQR=Q3-Q1
print("Median(Q2):",Q2)
print("Q1(25%):",Q1)
print("Q3(75%):",Q3)
print("Interquartile Range(IQR):",IQR)
```

```python
import numpy as np
marks=[42,48,51,55,58,60,63,67,70,72,75,78,80,83,85,88,90,92,95,98]
p1=np.percentile(marks,25)
p2=np.percentile(marks,50)
p3=np.percentile(marks,75)
print("25th Percentile:",p1)
print("50th Percentile:",p2)
print("75th Percentile:",p3)
```

```python
import numpy as np
import matplotlib.pyplot as plt
visitors=[-1,-30,22,25,27,30,32,35,36,38,40,42,45,50,52,55,58,90,95]
visitors.sort()
n=len(visitors)
Q1=int((n+1)/4)-1
Q3=int((3*(n+1))/4)-1
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3-1.5*IQR
cleared_data=[x for x in visitors if x<lower_limit or x>upper_limit]
plt.boxplot(cleared_data)
plt.show()
```
