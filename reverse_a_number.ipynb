```python
num=int(input("Enter a number:"))
rev=0
num1=num
if num<0:
  num=(-num)
while num > 0:
  dig=num%10
  rev=rev*10+dig
  num=num//10
if num1<0:
  rev=-rev
print("Reverse of entered number is:",rev)
```
