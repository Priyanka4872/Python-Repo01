```python
string=input()
res=" "
for i in range(len(string)):
  ch=string[i]
  if i==0 or i==string[i-1]==" ":
   if 'A'<=ch and ch<'Z':
    res+=chr(ord(ch)+32)
   else:
    res+=ch
  elif ch==" ":
     res+=" "
  else:
    if 'a'<=ch and ch<='z':
     res+=chr(ord(ch)-32)
    else:
      res+=ch
print(res)
```
